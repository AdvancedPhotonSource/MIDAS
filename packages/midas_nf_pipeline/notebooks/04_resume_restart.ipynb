# 04 · Resume & restart

Every pipeline run records its progress into a *pipeline H5* (one per
layer, named `<MicFileText>_pipeline.h5`). After each completed stage,
the H5 picks up a `pipeline_state/completed_stages/<i> = <stage_label>`
entry.

On the next run you have two options:

| Flag | Behaviour |
| --- | --- |
| `--resume <h5>` | auto-detect the first incomplete stage, skip everything before |
| `--restart-from <stage>` | force restart from a named stage |

Stage labels match the legacy `nf_MIDAS_Multiple_Resolutions.py`:

    loop_0_initial  loop_1_seeded  loop_1_unseeded  loop_1_merge
    loop_2_seeded   loop_2_unseeded  loop_2_merge   ...

## Resume via CLI

```bash
midas-nf-pipeline run params.txt --resume <pipeline_h5>
midas-nf-pipeline run params.txt --restart-from loop_2_unseeded
```

## Reset a pipeline (force re-run from a stage)

```python
from midas_nf_pipeline.state import PipelineH5
with PipelineH5('foo_pipeline.h5', 'nf_midas') as ph5:
    ph5.reset_from('loop_1_seeded', stage_order=[
        'loop_0_initial', 'loop_1_seeded', 'loop_1_unseeded', 'loop_1_merge'])
```
